```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import load_model
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import LSTM, Dropout, Dense
from tensorflow.keras.models import Sequential

```

#1st approach :Predicting All 96 Values at Once (Multi-Output Forecasting)
You use a past window of 96 time steps (e.g., one day) to predict the next 96 time steps (the entire next day).
The model outputs a vector of 96 future values in one shot.<br>



```python
def create_sequences_multistep(data, sequence_length, n_outputs):
        X, y = [], []
        for i in range(len(data) - sequence_length - n_outputs + 1):
            X.append(data[i:i + sequence_length])
            y.append(data[i + sequence_length:i + sequence_length + n_outputs])
        return np.array(X), np.array(y)
```

```python
def build_lstm_model(input_shape, n_outputs):
    model = Sequential([
        LSTM(64, return_sequences=True,input_shape=input_shape),
        Dropout(0.25),
        LSTM(32),
        Dropout(0.2),
        Dense(n_outputs)
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model
```

```python
def train_forecast_model(df, dir_name, sequence_length=96, n_outputs=96):
    print(f"\n📁 Processing directory: {dir_name}")

    # Step 1: Preprocessing
    df.set_index('timestamp', inplace=True)
    df = df[['storage_gb']]

    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(df.values).flatten()



    X, y = create_sequences_multistep(scaled_data, sequence_length, n_outputs)
    X = X.reshape((X.shape[0], X.shape[1], 1))

    # Train-test split
    split_idx = int(len(X) * 0.8)
    X_train, y_train = X[:split_idx], y[:split_idx]
    X_test, y_test = X[split_idx:], y[split_idx:]

    input_shape = (X_train.shape[1], X_train.shape[2])
    model = build_lstm_model(input_shape, n_outputs)

    # Step 3: Train
    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    history = model.fit(
        X_train, y_train,
        validation_split=0.2,
        epochs=100,
        batch_size=16,
        callbacks=[early_stop],
        verbose=0
    )

    # Step 4: Predictions & Metrics
    y_pred = model.predict(X_test)
    y_pred_original = scaler.inverse_transform(y_pred)
    y_test_original = scaler.inverse_transform(y_test)

    mae = mean_absolute_error(y_test_original[:, 95], y_pred_original[:, 95])
    rmse = np.sqrt(mean_squared_error(y_test_original[:, 95], y_pred_original[:, 95]))

    print(f"📊 {dir_name} - MAE (96th timestep): {mae:.2f} GB")
    print(f"📊 {dir_name} - RMSE (96th timestep): {rmse:.2f} GB")

    # Step 5: Plot
    plt.figure(figsize=(12, 5))
    plt.plot(y_test_original[:, 95], label='Actual')
    plt.plot(y_pred_original[:, 95], label='Predicted')
    plt.title(f"{dir_name.upper()} - Forecast for 96th Timestep (Next Day End)")
    plt.xlabel("Test Sample Index")
    plt.ylabel("Storage (GB)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    model.save(f"{dir_name}_1_day_forecast_model.h5")
```

```python
directories = {
    "info": info_df,
    "customer": customer_df,
    "scratch": scratch_df,
    "projects": projects_df
}

for dir_name, df in directories.items():
    train_forecast_model(df, dir_name)
```

```python
def evaluate_normalized_metrics(y_true, y_pred, label=""):
    # Ensure they are numpy arrays
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    # Select the 96th timestep (or change index if needed)
    y_true_96 = y_true[:, 95]
    y_pred_96 = y_pred[:, 95]

    # Compute metrics
    mae = mean_absolute_error(y_true_96, y_pred_96)
    rmse = np.sqrt(mean_squared_error(y_true_96, y_pred_96))
    data_range = y_true_96.max() - y_true_96.min()
    nrmse = rmse / data_range
    nmae = mae / data_range

    print(f"📁 {label} Metrics for Timestep 96:")
    print(f"   MAE   : {mae:.2f} GB")
    print(f"   RMSE  : {rmse:.2f} GB")
    print(f"   Normalized MAE  : {nmae:.4f}")
    print(f"   Normalized RMSE : {nrmse:.4f}")
    print("-" * 40)
```

```python
def evaluate_model_for_directory(df, dir_name, model_name):
    """
    Loads model, regenerates test data, and evaluates predictions for a given directory.
    """
    print(f"\n🔄 Evaluating model for directory: {dir_name}")
    df = df[['storage_gb']]

    # Preprocess and scale
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(df.values).flatten()

    # Sequence generation
    sequence_length = 96
    n_outputs = 96
    X, y = create_sequences_multistep(scaled_data, sequence_length, n_outputs)
    X = X.reshape((X.shape[0], X.shape[1], 1))

    # Split data
    split_idx = int(len(X) * 0.8)
    X_test, y_test = X[split_idx:], y[split_idx:]

    # Load model
    model = load_model(model_name,compile=False)

    # Predict
    y_pred = model.predict(X_test)
    y_pred_original = scaler.inverse_transform(y_pred)
    y_test_original = scaler.inverse_transform(y_test)

    # Evaluate

    evaluate_normalized_metrics(y_test_original, y_pred_original, label=dir_name)
```

```python
evaluate_model_for_directory(info_df, 'info', 'info_1_day_forecast_model.h5')
evaluate_model_for_directory(scratch_df, 'scratch', 'scratch_1_day_forecast_model.h5')
evaluate_model_for_directory(customer_df, 'customer', 'customer_1_day_forecast_model.h5')
evaluate_model_for_directory(projects_df, 'projects', 'projects_1_day_forecast_model.h5')
```

#2nd approach :Predicting Only the 96th Value (Single-Output Forecasting)
You still feed the past 96 time steps as input, but now the model predicts only the final value of the next day (i.e., the 96th time step ahead).The model returns a single scalar representing the storage at the end of next day.

```python
def create_sequences_singlestep(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i + sequence_length])
        y.append(data[i + sequence_length])  # Just one value
    return np.array(X), np.array(y)
```

```python
def build_lstm_model_singlestep(input_shape):
    model = Sequential([
        LSTM(64, return_sequences=True,input_shape=input_shape),
        Dropout(0.25),
        LSTM(32),
        Dropout(0.2),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model
```

```python
def train_single_step_forecast_model(df, dir_name, sequence_length=96):

    print(f"\n📁 Processing directory: {dir_name} (Single-Step Forecast)")

    # Step 1: Preprocessing
    df.set_index('timestamp', inplace=True)
    df = df[['storage_gb']]
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(df.values).flatten()


    X, y = create_sequences_singlestep(scaled_data, sequence_length)
    X = X.reshape((X.shape[0], X.shape[1], 1))  # LSTM expects 3D
    y = y.reshape(-1, 1)

    # Step 3: Train-test split
    split_idx = int(len(X) * 0.8)
    X_train, y_train = X[:split_idx], y[:split_idx]
    X_test, y_test = X[split_idx:], y[split_idx:]

    input_shape = (X_train.shape[1], X_train.shape[2])
    model = build_lstm_model_singlestep(input_shape)

    # Step 5: Training
    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    history = model.fit(
        X_train, y_train,
        validation_split=0.2,
        epochs=100,
        batch_size=16,
        callbacks=[early_stop],
        verbose=0
    )

    # Step 6: Prediction
    y_pred = model.predict(X_test)
    y_pred_original = scaler.inverse_transform(y_pred)
    y_test_original = scaler.inverse_transform(y_test)

    # Step 7: Metrics
    mae = mean_absolute_error(y_test_original, y_pred_original)
    rmse = np.sqrt(mean_squared_error(y_test_original, y_pred_original))
    data_range = y_test_original.max() - y_test_original.min()
    nmae = mae / data_range
    nrmse = rmse / data_range

    print(f"📊 {dir_name.upper()} - MAE: {mae:.2f} GB")
    print(f"📊 {dir_name.upper()} - RMSE: {rmse:.2f} GB")
    print(f"📊 {dir_name.upper()} - Normalized MAE: {nmae:.4f}")
    print(f"📊 {dir_name.upper()} - Normalized RMSE: {nrmse:.4f}")
    print("-" * 50)

    # Step 8: Plot
    plt.figure(figsize=(12, 5))
    plt.plot(y_test_original, label='Actual')
    plt.plot(y_pred_original, label='Predicted')
    plt.title(f"{dir_name.upper()} - Single-Step Forecast (Next Day End)")
    plt.xlabel("Test Sample Index")
    plt.ylabel("Storage (GB)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Step 9: Save model
    model.save(f"{dir_name}_1_day_single_step_forecast_model.keras")
```

```python
directories = {
    "info": info_df,
    "customer": customer_df,
    "scratch": scratch_df,
    "projects": projects_df
}

for dir_name, df in directories.items():
    train_single_step_forecast_model(df.copy(), dir_name)
```

#Conclusion
